```python
#Handing missing values(using  mean and scikit learn)
from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values = np.nan,strategy = 'mean')#fill mesing values in age and salary
ds[['Age','Salary']]=si.fit_transform(ds[['Age','Salary']])
```

```python
ds
```

```python
ds["Country"].unique()
```

```python
#Encoding Categorical Column(using oneHotEncoding)
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output = False)
encoded_data=ohe.fit_transform(ds[['Country']])
encoded_data
```

```python
#Creating columns
column_names = ohe.get_feature_names_out(['Country'])
encoded_df = pd.DataFrame(encoded_data,columns =column_names)#convert array into dataset
encoded_df
```

```python
column_names#display title in file 
```

```python
#joing original dataset and new dataframe

ds = pd.concat([ds,encoded_df],axis = 1)
ds
```

```python
#Drop actual categorical country column

ds = ds.drop(columns=['Country'])#drop country columnn
ds
```

```python
#label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
ds['Purchased']=le.fit_transform(ds['Purchased'])
```

```python
ds
```

```python
#Feature Scaling(Reduce large data values to its smaller datavalues)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
ds[['Age','Salary']]=sc.fit_transform(ds[['Age','Salary']])
```
